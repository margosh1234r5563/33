```python
# @title ПОЛНЫЙ ИСПРАВЛЕННЫЙ КОД С ИНТЕГРИРОВАННЫМИ КОРРЕКТИРОВКАМИ

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from scipy.stats import norm
import warnings
warnings.filterwarnings('ignore')

from google.colab import files
import io

print("ОПТИМИЗАЦИОННАЯ МОДЕЛЬ УПРАВЛЕНИЯ ЗАПАСАМИ S7 AIRLINES")
print("Версия: 10.0 (С ИСПРАВЛЕНИЕМ МАСШТАБА)")
print("Дата расчета: 2026 год")
print("\n📥 ЗАГРУЗКА ВСЕХ ДАННЫХ...")

# Загружаем ВСЕ файлы
print("\n1. Загрузите category_metrics_full_two_versions.csv:")
uploaded_metrics = files.upload()
metrics_filename = list(uploaded_metrics.keys())[0]
metrics_df = pd.read_csv(io.BytesIO(uploaded_metrics[metrics_filename]))

print("\n2. Загрузите final_orders_train.csv:")
uploaded_orders = files.upload()
orders_filename = list(uploaded_orders.keys())[0]
orders_df = pd.read_csv(io.BytesIO(uploaded_orders[orders_filename]))

print("\n3. Загрузите final_consumtion_train.csv:")
uploaded_cons = files.upload()
cons_filename = list(uploaded_cons.keys())[0]
consumption_df = pd.read_csv(io.BytesIO(uploaded_cons[cons_filename]))

# Преобразуем даты
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])
orders_df['delivery_date'] = pd.to_datetime(orders_df['delivery_date'])
orders_df['delivery_time'] = (orders_df['delivery_date'] - orders_df['order_date']).dt.days

consumption_df['consumtion_date'] = pd.to_datetime(consumption_df['consumtion_date'])
consumption_df['year'] = consumption_df['consumtion_date'].dt.year
consumption_df['month'] = consumption_df['consumtion_date'].dt.month

print(f"\n✅ Все данные загружены:")
print(f"   Метрики: {len(metrics_df)} категорий")
print(f"   Заказы: {len(orders_df)} записей")
print(f"   Потребление: {len(consumption_df)} записей")
```

```python
# @title ШАГ 1: ИСПОЛЬЗОВАНИЕ ВАШИХ МЕТРИК КАК БАЗЫ

print("ШАГ 1: ИСПОЛЬЗОВАНИЕ ВАШИХ МЕТРИК КАК БАЗЫ")
# Преобразуем метрики в удобный формат
category_metrics = {}
for _, row in metrics_df.iterrows():
    cat = int(row['product_category'])
    category_metrics[cat] = {
        'demand_mean': row['demand_mean'],
        'demand_std': row['demand_std'],
        'demand_cv': row['demand_cv_category_agg'],
        'lead_time_mean': row['lead_time_mean'],
        'lead_time_std': row['lead_time_std'],
        'lead_time_cv': row['lead_time_cv_category_agg'],
        'price_mean': row['price_mean'],
        'price_std': row['price_std'],
        'price_cv': row['price_cv_category_agg'],
        'shortage_rate': row['shortage_rate']
    }

print(f"\n📊 ЗАГРУЖЕННЫЕ МЕТРИКИ (ИЗ ВАШЕГО ФАЙЛА):")
for cat in sorted(category_metrics.keys()):
    m = category_metrics[cat]
    annual_demand = m['demand_mean'] * 365
    print(f"\nКатегория {cat}:")
    print(f"   Спрос: {m['demand_mean']:.2f} ед/день → {annual_demand:.0f} ед/год")
    print(f"   Поставки: {m['lead_time_mean']:.1f} ± {m['lead_time_std']:.1f} дней")
    print(f"   Цена: {m['price_mean']:,.0f} руб")
```

```python
# @title ШАГ 2: ОПРЕДЕЛЕНИЕ СТРАТЕГИЙ

print("ШАГ 2: ОПРЕДЕЛЕНИЕ СТРАТЕГИЙ ПО КАТЕГОРИЯМ")

strategies = {
    0: {
        'abc_xyz': 'AZ',
        'name': 'ROP с максимальным потреблением',
        'service_level': 0.98,
        'z_score': norm.ppf(0.98),
        'logic': 'AZ - нестабильный спрос, высокий риск AOG'
    },
    1: {
        'abc_xyz': 'AX',
        'name': 'JIC с мониторингом цен',
        'service_level': 0.98,
        'z_score': norm.ppf(0.98),
        'logic': 'AX - очень дорогие, волатильная цена'
    },
    2: {
        'abc_xyz': 'BY',
        'name': 'JIC с повышенным запасом',
        'service_level': 0.95,
        'z_score': norm.ppf(0.95),
        'logic': 'BY - нестабильный спрос и поставки'
    },
    3: {
        'abc_xyz': 'CZ',
        'name': 'Еженедельные заказы',
        'service_level': 0.99,
        'z_score': norm.ppf(0.99),
        'logic': 'CZ - критичные расходники, хаотичный спрос'
    },
    4: {
        'abc_xyz': 'AZ',
        'name': 'ROP с максимальным потреблением',
        'service_level': 0.98,
        'z_score': norm.ppf(0.98),
        'logic': 'AZ - высокий вклад, нестабильный спрос'
    }
}

print("\n📌 ОПРЕДЕЛЕНЫ СЛЕДУЮЩИЕ СТРАТЕГИИ:")
for cat in sorted(strategies.keys()):
    s = strategies[cat]
    print(f"\nКатегория {cat} ({s['abc_xyz']}): {s['name']} (уровень {s['service_level']*100:.0f}%)")
```

```python
# @title ШАГ 3: РАСЧЕТ ПАРАМЕТРОВ УПРАВЛЕНИЯ ДЛЯ КАТЕГОРИЙ

print("ШАГ 3: РАСЧЕТ ПАРАМЕТРОВ УПРАВЛЕНИЯ ЗАПАСАМИ")

def calculate_category_parameters(cat, metrics, strategy):
    """
    РАСЧЕТ ПАРАМЕТРОВ ДЛЯ КАТЕГОРИИ
    """

    print(f"\n{'='*60}")
    print(f"РАСЧЕТ ДЛЯ КАТЕГОРИИ {cat} ({strategy['abc_xyz']})")
    print(f"{'='*60}")

    # Исходные данные
    D = metrics['demand_mean']
    σ_D = metrics['demand_std']
    L = metrics['lead_time_mean']
    σ_L = metrics['lead_time_std']
    P = metrics['price_mean']
    z = strategy['z_score']

    # Максимальное потребление
    if cat in [0, 2, 3, 4]:
        D_max = D + 2 * σ_D
    else:
        D_max = D

    # Расчет страхового запаса
    demand_uncertainty = L * (σ_D ** 2)
    lead_time_uncertainty = (D ** 2) * (σ_L ** 2)
    total_uncertainty = np.sqrt(demand_uncertainty + lead_time_uncertainty)
    safety_stock = z * total_uncertainty

    # Точка перезаказа
    demand_during_lead = D_max * L
    rop = demand_during_lead + safety_stock

    # Размер заказа
    if cat == 1:
        order_qty = 2
        interval = 30
        orders_per_year = 12
    elif cat == 3:
        order_qty = max(1, round(D * 14))
        interval = 7
        orders_per_year = 52
    else:
        order_qty = max(1, round(D * 30))
        interval = 30
        orders_per_year = 12

    # Корректировка для категории 2
    if cat == 2:
        safety_stock *= 1.3
        rop *= 1.2
        order_qty = round(order_qty * 1.3)

    params = {
        'category': cat,
        'abc_xyz': strategy['abc_xyz'],
        'strategy': strategy['name'],
        'service_level': strategy['service_level'],
        'safety_stock': max(1, round(safety_stock)),
        'rop': max(1, round(rop)),
        'order_quantity': max(1, round(order_qty)),
        'interval_days': interval,
        'orders_per_year': orders_per_year,
        'price': P,
        'daily_demand': D
    }

    print(f"\n✅ ПАРАМЕТРЫ КАТЕГОРИИ:")
    print(f"   Страховой запас: {params['safety_stock']} ед.")
    print(f"   ROP: {params['rop']} ед.")
    print(f"   Размер заказа: {params['order_quantity']} ед.")
    print(f"   Заказов в год: {params['orders_per_year']}")

    return params

# Рассчитываем параметры для категорий
category_params_list = []
for cat in sorted(category_metrics.keys()):
    params = calculate_category_parameters(cat, category_metrics[cat], strategies[cat])
    category_params_list.append(params)
```

```python
# @title ШАГ 4: ПРОГНОЗ СПРОСА НА 2026 ГОД

print("ШАГ 4: ПРОГНОЗ СПРОСА НА 2026 ГОД")

months = ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн',
          'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек']

# Анализ сезонности
seasonality = {}
for cat in sorted(category_metrics.keys()):
    cat_cons = consumption_df[consumption_df['product_category'] == cat]
    if len(cat_cons) > 0:
        monthly_avg = cat_cons.groupby('month')['qty'].sum()
        overall_avg = monthly_avg.mean()
        if overall_avg > 0:
            seasonality[cat] = (monthly_avg / overall_avg).to_dict()
        else:
            seasonality[cat] = {m: 1.0 for m in range(1, 13)}
    else:
        seasonality[cat] = {m: 1.0 for m in range(1, 13)}

# Прогноз спроса
demand_forecasts = {}

for cat in sorted(category_metrics.keys()):
    D = category_metrics[cat]['demand_mean']
    annual_demand = D * 365
    base_monthly = annual_demand / 12

    forecast = []
    for month in range(1, 13):
        seas_factor = seasonality[cat].get(month, 1.0)
        demand = base_monthly * seas_factor
        forecast.append({
            'Месяц': months[month-1],
            'Спрос': round(demand)
        })

    # Корректировка до годовой суммы
    total = sum(f['Спрос'] for f in forecast)
    factor = annual_demand / total
    for f in forecast:
        f['Спрос'] = round(f['Спрос'] * factor)

    demand_forecasts[cat] = forecast

    print(f"\n📊 Категория {cat}: годовой спрос = {annual_demand:.0f} ед.")
```

```python
# @title ШАГ 5: РАСЧЕТ ПАРАМЕТРОВ ДЛЯ КАЖДОГО ТОВАРА (С ИСПРАВЛЕНИЕМ МАСШТАБА)

print("ШАГ 5: РАСЧЕТ ПАРАМЕТРОВ ДЛЯ КАЖДОГО ТОВАРА")

# Получаем все уникальные товары
all_products = consumption_df[['product_id', 'product_category']].drop_duplicates()
print(f"\n📦 ВСЕГО ТОВАРОВ: {len(all_products)}")

def calculate_product_parameters_safe(product_id, category):
    """
    БЕЗОПАСНЫЙ РАСЧЕТ ПАРАМЕТРОВ ТОВАРА С КОНТРОЛЕМ МАСШТАБА
    """

    # Данные по товару
    prod_cons = consumption_df[consumption_df['product_id'] == product_id]
    prod_orders = orders_df[orders_df['product_id'] == product_id]

    # Параметры категории
    cat_params = [p for p in category_params_list if p['category'] == category][0]
    cat_metric = category_metrics[category]

    # Спрос товара
    if len(prod_cons) > 0:
        daily_cons = prod_cons.groupby('consumtion_date')['qty'].sum()
        D_product = daily_cons.mean()
    else:
        D_product = cat_metric['demand_mean']

    # Цена товара
    if len(prod_orders) > 0:
        price = (prod_orders['amount'] / prod_orders['qty']).mean()
    else:
        price = cat_metric['price_mean']

    # !!! ВАЖНО: Ограничиваем размер заказа разумными пределами
    # Месячный спрос товара
    monthly_demand = D_product * 30

    # Базовый размер заказа от категории
    base_order = cat_params['order_quantity']

    # Корректируем под товар, но с жесткими ограничениями
    if D_product > 0:
        # Отношение спроса товара к среднему по категории
        cat_avg_demand = cat_metric['demand_mean']
        if cat_avg_demand > 0:
            scale = D_product / cat_avg_demand
        else:
            scale = 1.0
    else:
        scale = 1.0

    # Ограничиваем scale разумными пределами (0.2 - 5)
    scale = max(0.2, min(5.0, scale))

    # Рассчитываем размер заказа
    if category == 1:
        # Для очень дорогих - минимальные заказы
        order_qty = max(1, min(5, round(monthly_demand * 0.5)))
    elif category == 3:
        # Для расходников - недельный запас
        order_qty = max(1, round(D_product * 7))
    else:
        # Для остальных - месячный запас, но не более 500 ед.
        order_qty = max(1, min(500, round(monthly_demand)))

    # Страховой запас с ограничением
    safety_stock = max(1, min(1000, round(cat_params['safety_stock'] * scale)))

    # ROP с ограничением
    rop = max(1, min(2000, round(cat_params['rop'] * scale)))

    return {
        'product_id': product_id,
        'category': category,
        'abc_xyz': cat_params['abc_xyz'],
        'strategy': cat_params['strategy'],
        'daily_demand': round(D_product, 2),
        'monthly_demand': round(monthly_demand),
        'price': round(price),
        'order_quantity': order_qty,
        'orders_per_year': cat_params['orders_per_year'],
        'safety_stock': safety_stock,
        'rop': rop,
        'scale_factor': round(scale, 2)
    }

# Рассчитываем для всех товаров
product_params_list = []

for _, row in all_products.iterrows():
    product_id = row['product_id']
    category = row['product_category']

    params = calculate_product_parameters_safe(product_id, category)
    product_params_list.append(params)

product_params_df = pd.DataFrame(product_params_list)

print("\n📊 ПАРАМЕТРЫ ПО ТОВАРАМ (ПЕРВЫЕ 20):")
display_cols = ['product_id', 'category', 'daily_demand', 'price', 'order_quantity', 'safety_stock', 'rop']
print(product_params_df[display_cols].head(20).to_string(index=False))

# Проверка на аномалии
print("\n🔍 ПРОВЕРКА НА АНОМАЛИИ:")
print(f"   Макс. цена: {product_params_df['price'].max():,.0f} руб")
print(f"   Макс. размер заказа: {product_params_df['order_quantity'].max()} ед.")
print(f"   Средний размер заказа: {product_params_df['order_quantity'].mean():.0f} ед.")

# Расчет годовых затрат
product_params_df['annual_volume'] = product_params_df['order_quantity'] * product_params_df['orders_per_year']
product_params_df['annual_cost'] = product_params_df['annual_volume'] * product_params_df['price']

total_annual_cost = product_params_df['annual_cost'].sum()
print(f"\n💰 ОБЩАЯ ГОДОВАЯ СТОИМОСТЬ ПО ВСЕМ ТОВАРАМ: {total_annual_cost:,.0f} руб")
print(f"   {total_annual_cost/1e9:.2f} млрд руб")
```

```python
# @title ШАГ 6: ФОРМИРОВАНИЕ ПЛАНА ЗАКУПОК ПО ТОВАРАМ

print("ШАГ 6: ФОРМИРОВАНИЕ ПЛАНА ЗАКУПОК ПО ТОВАРАМ")

procurement_plan = []

for idx, product in product_params_df.iterrows():
    cat = product['category']
    product_id = product['product_id']

    if cat not in demand_forecasts:
        continue

    cat_forecast = demand_forecasts[cat]

    # Распределяем заказы по месяцам
    for month_idx, month_data in enumerate(cat_forecast):
        month_name = month_data['Месяц']

        # Количество заказов в месяце
        if cat == 3:
            weeks_in_month = [4, 4, 5, 4, 4, 5, 4, 4, 5, 4, 4, 5]
            n_orders = weeks_in_month[month_idx]
        else:
            n_orders = max(1, round(product['orders_per_year'] / 12))

        # Генерируем заказы
        for order_num in range(1, n_orders + 1):
            # Равномерно распределяем по месяцу
            day = round((order_num / (n_orders + 1)) * 28)
            day = max(1, min(28, day))

            month_num = month_idx + 1
            order_date = f"2026-{month_num:02d}-{day:02d}"

            # Дата поставки
            lead_time = category_metrics[cat]['lead_time_mean']
            delivery_date = (pd.to_datetime(order_date) + timedelta(days=int(lead_time))).strftime('%Y-%m-%d')

            # Стоимость заказа
            order_cost = product['order_quantity'] * product['price']

            procurement_plan.append({
                'product_id': product_id,
                'category': cat,
                'abc_xyz': product['abc_xyz'],
                'strategy': product['strategy'],
                'order_date': order_date,
                'delivery_date': delivery_date,
                'quantity': product['order_quantity'],
                'price_per_unit': product['price'],
                'order_cost_rub': order_cost,
                'month': month_name
            })

# Сортируем по дате
plan_df = pd.DataFrame(procurement_plan)
plan_df = plan_df.sort_values(['order_date', 'category', 'product_id'])

print(f"\n📊 СФОРМИРОВАНО ЗАПИСЕЙ: {len(plan_df)}")
print("\nПЕРВЫЕ 20 СТРОК ПЛАНА:")
print(plan_df.head(20).to_string(index=False))
```

```python
# @title СОЗДАНИЕ CSV ФАЙЛА С ПЛАНОМ ЗАКУПОК ПО ТОВАРАМ НА 2026 ГОД (ИСПРАВЛЕНО)

print("СОЗДАНИЕ CSV ФАЙЛА С ПЛАНОМ ЗАКУПОК ПО ТОВАРАМ НА 2026 ГОД")

# ============================================
# 1. ПОЛУЧАЕМ СПИСОК ВСЕХ ТОВАРОВ
# ============================================

# Получаем уникальные товары из consumption_df
all_products = consumption_df[['product_id', 'product_category']].drop_duplicates()

print(f"\n📦 ВСЕГО ТОВАРОВ: {len(all_products)}")
print("\nРАСПРЕДЕЛЕНИЕ ПО КАТЕГОРИЯМ:")
print(all_products['product_category'].value_counts().sort_index())

# ============================================
# 2. РАСЧЕТ ПАРАМЕТРОВ ДЛЯ КАЖДОГО ТОВАРА
# ============================================

def calculate_product_parameters(product_id, category, category_metrics, strategies, params_list):
    """
    РАСЧЕТ ПАРАМЕТРОВ ДЛЯ КОНКРЕТНОГО ТОВАРА
    """

    # Данные по товару
    prod_cons = consumption_df[consumption_df['product_id'] == product_id]
    prod_orders = orders_df[orders_df['product_id'] == product_id]

    # Берем базовые параметры из категории
    cat_metric = category_metrics[category]
    D_cat = cat_metric['demand_mean']
    price_cat = cat_metric['price_mean']

    # Параметры из стратегии категории
    strategy = strategies[category]
    cat_params = [p for p in params_list if p['category'] == category][0]

    # Корректируем под конкретный товар
    if len(prod_cons) > 0:
        daily_cons = prod_cons.groupby('consumtion_date')['qty'].sum()
        D_prod = daily_cons.mean()
        # Если есть данные по товару, используем их
        if D_prod > 0:
            D = D_prod
        else:
            D = D_cat
    else:
        D = D_cat

    # Цена товара (если есть в заказах)
    if len(prod_orders) > 0:
        price = (prod_orders['amount'] / prod_orders['qty']).mean()
    else:
        price = price_cat

    # Масштабируем размер заказа под товар
    # Если спрос товара отличается от среднего по категории
    if D_cat > 0:
        scale_factor = D / D_cat
    else:
        scale_factor = 1.0

    order_qty = max(1, round(cat_params['order_quantity'] * scale_factor))
    safety_stock = max(1, round(cat_params['safety_stock'] * scale_factor))
    rop = max(1, round(cat_params['rop'] * scale_factor))

    return {
        'product_id': product_id,
        'category': category,
        'abc_xyz': cat_params['abc_xyz'],
        'strategy': cat_params['strategy'],
        'daily_demand': round(D, 2),
        'price': round(price),
        'order_quantity': order_qty,
        'orders_per_year': cat_params['orders_per_year'],
        'safety_stock': safety_stock,
        'rop': rop,
        'scale_factor': round(scale_factor, 2)
    }

# Рассчитываем параметры для всех товаров
product_params_list = []

for _, row in all_products.iterrows():
    product_id = row['product_id']
    category = row['product_category']

    product_params = calculate_product_parameters(
        product_id,
        category,
        category_metrics,
        strategies,
        params_list
    )
    product_params_list.append(product_params)

product_params_df = pd.DataFrame(product_params_list)
print("\n📊 ПАРАМЕТРЫ ПО ТОВАРАМ (ПЕРВЫЕ 20):")
print(product_params_df.head(20).to_string(index=False))

# ============================================
# 3. ФОРМИРОВАНИЕ ДЕТАЛЬНОГО ПЛАНА ЗАКУПОК ПО ТОВАРАМ
# ============================================

months = ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн',
          'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек']

procurement_plan_by_product = []

# ИСПРАВЛЕНИЕ: Используем iterrows() вместо прямого перебора
for idx, product in product_params_df.iterrows():
    cat = product['category']
    product_id = product['product_id']

    # Прогноз спроса для категории
    if cat in forecasts:
        cat_forecast = forecasts[cat]

        # Доля товара в общем спросе категории
        cat_total_demand = category_metrics[cat]['demand_mean'] * 365
        if cat_total_demand > 0:
            product_demand_share = product['daily_demand'] * 365 / cat_total_demand
        else:
            product_demand_share = 1 / len(product_params_df[product_params_df['category'] == cat])

        for month_idx, month_data in enumerate(cat_forecast):
            month_name = month_data['Месяц']
            month_demand = month_data['Спрос']

            # Спрос на этот товар в месяце
            product_month_demand = round(month_demand * product_demand_share)

            # Количество заказов в месяце для этого товара
            if cat == 3:
                # Для категории 3 (еженедельные)
                weeks_in_month = [4, 4, 5, 4, 4, 5, 4, 4, 5, 4, 4, 5]
                n_orders = weeks_in_month[month_idx]
            else:
                n_orders = max(1, round(product['orders_per_year'] / 12))

            # Генерируем заказы
            for order_num in range(1, n_orders + 1):
                # Дата заказа (равномерно по месяцу)
                days_in_month = 30
                day_of_month = round((order_num / (n_orders + 1)) * days_in_month)
                day_of_month = max(1, min(28, day_of_month))

                month_num = month_idx + 1
                order_date = f"2026-{month_num:02d}-{day_of_month:02d}"

                # Дата поставки
                lead_time = category_metrics[cat]['lead_time_mean']
                delivery_date = (pd.to_datetime(order_date) + timedelta(days=int(lead_time))).strftime('%Y-%m-%d')

                # Стоимость заказа
                order_cost = product['order_quantity'] * product['price']

                procurement_plan_by_product.append({
                    'product_id': product_id,
                    'category': cat,
                    'abc_xyz': product['abc_xyz'],
                    'strategy': product['strategy'],
                    'month': month_name,
                    'order_date': order_date,
                    'delivery_date': delivery_date,
                    'quantity': product['order_quantity'],
                    'price_per_unit': product['price'],
                    'order_cost_rub': order_cost,
                    'month_demand_forecast': product_month_demand
                })

# Сортируем по дате
plan_df = pd.DataFrame(procurement_plan_by_product)
if len(plan_df) > 0:
    plan_df = plan_df.sort_values(['order_date', 'category', 'product_id'])

print(f"\n📊 СФОРМИРОВАНО ЗАПИСЕЙ В ПЛАНЕ: {len(plan_df)}")
if len(plan_df) > 0:
    print("\nПЕРВЫЕ 20 СТРОК ПЛАНА ПО ТОВАРАМ:")
    print(plan_df.head(20).to_string(index=False))
else:
    print("❌ План пуст!")

# ============================================
# 4. СОХРАНЕНИЕ ПОЛНОГО ПЛАНА ПО ТОВАРАМ
# ============================================

if len(plan_df) > 0:
    full_plan_filename = 's7_procurement_plan_by_product_2026.csv'
    plan_df.to_csv(full_plan_filename, index=False, encoding='utf-8-sig')
    print(f"\n✅ Полный план по товарам сохранен: '{full_plan_filename}'")

    # ============================================
    # 5. УПРОЩЕННАЯ ВЕРСИЯ
    # ============================================

    simple_plan = plan_df[[
        'order_date',
        'product_id',
        'category',
        'quantity',
        'order_cost_rub'
    ]].copy()

    simple_plan.columns = ['date', 'product_id', 'category', 'quantity', 'cost_rub']

    simple_filename = 's7_procurement_plan_simple_2026.csv'
    simple_plan.to_csv(simple_filename, index=False, encoding='utf-8-sig')
    print(f"✅ Упрощенный план по товарам сохранен: '{simple_filename}'")

    # ============================================
    # 6. СТАТИСТИКА
    # ============================================

    print("СТАТИСТИКА ПЛАНА ЗАКУПОК ПО ТОВАРАМ")

    # Статистика по категориям
    cat_stats = plan_df.groupby('category').agg({
        'quantity': 'sum',
        'order_cost_rub': 'sum',
        'product_id': 'nunique',
        'order_date': 'count'
    }).round(0)

    cat_stats.columns = ['всего_единиц', 'общая_стоимость_руб', 'кол-во_товаров', 'кол-во_заказов']
    cat_stats['стоимость_млн_руб'] = (cat_stats['общая_стоимость_руб'] / 1_000_000).round(2)

    print("\n📊 СТАТИСТИКА ПО КАТЕГОРИЯМ:")
    print(cat_stats.to_string())

    # Общая статистика
    total_orders = len(plan_df)
    total_units = plan_df['quantity'].sum()
    total_cost = plan_df['order_cost_rub'].sum()
    unique_products = plan_df['product_id'].nunique()

    print(f"\n📈 ОБЩАЯ СТАТИСТИКА:")
    print(f"   Всего товаров: {unique_products}")
    print(f"   Всего заказов: {total_orders}")
    print(f"   Всего единиц: {total_units:,.0f}")
    print(f"   Общая стоимость: {total_cost:,.0f} руб")
    print(f"   Общая стоимость: {total_cost/1_000_000:,.2f} млн руб")

    # ============================================
    # 7. СКАЧИВАНИЕ ФАЙЛОВ
    # ============================================

    from google.colab import files

    print("СКАЧИВАНИЕ ФАЙЛОВ")

    files.download(full_plan_filename)
    print(f"✅ {full_plan_filename} скачан")

    files.download(simple_filename)
    print(f"✅ {simple_filename} скачан")

    print("✅ ПЛАН ЗАКУПОК ПО ТОВАРАМ УСПЕШНО СОЗДАН И СКАЧАН")
else:
    print("\n❌ ОШИБКА: План не сформирован. Проверьте наличие данных в forecasts")
```

```python
# @title ЭКСТРЕННОЕ ИСПРАВЛЕНИЕ МАСШТАБА

print("ЭКСТРЕННОЕ ИСПРАВЛЕНИЕ МАСШТАБА ЦЕН И ЗАКАЗОВ")

# Анализируем проблемные продукты
problematic_products = []

for _, product in product_params_df.iterrows():
    annual_cost = product['order_quantity'] * product['price'] * product['orders_per_year']
    if annual_cost > 1e12:  # больше 1 триллиона
        problematic_products.append({
            'product_id': product['product_id'],
            'category': product['category'],
            'price': product['price'],
            'order_qty': product['order_quantity'],
            'annual_cost': annual_cost
        })

if problematic_products:
    print("\n⚠️ НАЙДЕНЫ ПРОБЛЕМНЫЕ ТОВАРЫ С ЗАВЫШЕННЫМИ ЗНАЧЕНИЯМИ:")
    prob_df = pd.DataFrame(problematic_products)
    print(prob_df.to_string())

    # ИСПРАВЛЕНИЕ 1: Ограничиваем размер заказа разумными пределами
    print("\n🔧 ПРИМЕНЯЕМ КОРРЕКТИРОВКИ...")

    for idx, row in prob_df.iterrows():
        prod_id = row['product_id']
        cat = row['category']

        # Находим индекс продукта в product_params_df
        prod_idx = product_params_df[product_params_df['product_id'] == prod_id].index[0]

        # Исправление 1: Ограничиваем размер заказа
        old_qty = product_params_df.loc[prod_idx, 'order_quantity']

        # Новый разумный размер заказа (месячный спрос)
        D = product_params_df.loc[prod_idx, 'daily_demand']
        new_qty = max(1, round(D * 30))

        product_params_df.loc[prod_idx, 'order_quantity'] = new_qty

        # Исправление 2: Проверяем цену
        if product_params_df.loc[prod_idx, 'price'] > 10_000_000:
            print(f"   Товар {prod_id}: цена {product_params_df.loc[prod_idx, 'price']:,.0f} руб (очень высокая)")

        print(f"   Товар {prod_id}: размер заказа {old_qty} → {new_qty} ед.")

    print("\n✅ Корректировки применены")

    # Пересчитываем параметры после исправления
    product_params_df['annual_order_volume'] = product_params_df['order_quantity'] * product_params_df['orders_per_year']
    product_params_df['annual_cost'] = product_params_df['annual_order_volume'] * product_params_df['price']

    print("\n📊 НОВЫЕ ПАРАМЕТРЫ (ТОП-10 ПО СТОИМОСТИ):")
    top10 = product_params_df.nlargest(10, 'annual_cost')[['product_id', 'category', 'daily_demand', 'price', 'order_quantity', 'annual_cost']]
    print(top10.to_string(index=False))

    total_annual_cost = product_params_df['annual_cost'].sum()
    print(f"\n💰 ОБЩАЯ ГОДОВАЯ СТОИМОСТЬ ПОСЛЕ КОРРЕКТИРОВКИ: {total_annual_cost:,.0f} руб ({total_annual_cost/1e9:.2f} млрд руб)")
```
